```python
import matplotlib.pyplot as plt
import numpy as np

def scatter_plot_landing_times(data):
    # Split the data into lines
    lines = data.strip().split('\n')
    
    # Initialize lists to store landing times for each node
    node_20_times = []
    node_22_times = []
    
    # Parse the data
    for line in lines:
        parts = line.split()
        node = int(parts[0])
        
        if node == 20:
            node_20_times.extend([float(parts[i+1]) for i in range(1, len(parts), 2)])
        elif node == 22:
            node_22_times.extend([float(parts[i+1]) for i in range(1, len(parts), 2)])
    
    # Convert landing times to hours of the day
    def to_hour_of_day(time):
        return (time / 3600) % 24

    node_20_hours = [to_hour_of_day(t) for t in node_20_times]
    node_22_hours = [to_hour_of_day(t) for t in node_22_times]

    # Create the scatter plot
    plt.figure(figsize=(12, 4))
    plt.scatter(node_20_hours, np.zeros_like(node_20_hours), label='Node 20', alpha=0.6)
    plt.scatter(node_22_hours, np.ones_like(node_22_hours), label='Node 22', alpha=0.6)
    
    plt.xlabel('Hour of the Day')
    plt.title('Landing Times for Nodes 20 and 22 (24-hour cycle)')
    plt.legend()

    # Remove y-axis ticks and labels
    plt.yticks([])

    # Set x-axis ticks and limits
    plt.xticks(range(0, 25, 1))
    plt.xlim(0, 24)

    # Remove grid
    plt.grid(False)

    plt.tight_layout()
    plt.show()

    # Plot the histogram
    plt.figure(figsize=(12, 4))
    bins = np.arange(0, 25, 1)  # 24 bins, one for each hour, plus right edge
    plt.hist(node_20_hours, bins=bins, edgecolor='black', alpha=0.7, label='Node 20')
    plt.hist(node_22_hours, bins=bins, edgecolor='red', alpha=0.7, label='Node 22')
    
    plt.xlabel('Hour of the Day')
    plt.ylabel('Frequency')
    plt.title('Histogram of Landing Times for Nodes 20 and 22 (24-hour cycle)')
    plt.xticks(range(0, 25, 2))  # Show even hours on x-axis
    plt.legend()
    plt.tight_layout()
    plt.show()

# Use the function
scatter_plot_landing_times(data)
```

```python
# Read the content of the new file
with open('RESULT/ZGSZ_PMSZGSZ_PMS_py_eliminate0_outx1.5_RWY.res', 'r') as file:
    data = file.read()

# Use the function
scatter_plot_landing_times(data)
```
