```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
file_name = 'RESULT/ZGSZ_STDZGSZ_STD_FEATURES.res'
```

```python
# Load the CSV file into a pandas DataFrame
df_15x = pd.read_csv(file_name, header=None, names=['Temperature', 'OverallObjectiveValue', 'NumberOfConflicts', 'EvaluationOfNodes', 'EvaluationOfLinks', 'TotalDelay', 'SpdChange', 'RtChange'], delimiter=' ', index_col=False)
# df_2x = pd.read_csv('ZGSZ_PMSZGSZ_PMS_py_eliminate0_outx2_FEATURES.res', header=None, names=['Temperature', 'OverallObjectiveValue', 'NumberOfConflicts', 'EvaluationOfNodes', 'EvaluationOfLinks', 'TotalDelay', 'SpdChange', 'RtChange'], delimiter=' ', index_col=False)
```

# Plot the OverallObjectiveValue

```python
# Plot the OverallObjectiveValue
plt.figure(figsize=(12, 6))
plt.plot(df_15x.index, df_15x['OverallObjectiveValue'])
# plt.plot(df_2x.index, df_2x['OverallObjectiveValue'])
plt.title('Overall Objective Value over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Overall Objective Value')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.grid(True)
plt.show()
```

# Plot the NumberOfConflicts

```python
plt.figure(figsize=(12, 6))
plt.plot(df_15x.index, df_15x['NumberOfConflicts'])
# plt.plot(df_2x.index, df_2x['NumberOfConflicts'])
plt.title('Number of Conflicts over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Number of Conflicts')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.grid(True)
plt.show()
```
